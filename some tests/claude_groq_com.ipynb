```python
import anthropic
import os
import PyPDF2
import groq
_claude=anthropic.Anthropic(api_key=os.environ['ANTHROPIC_API_KEY'])
_groq=groq.Groq(api_key=os.environ.get('GROQ_API_KEY'))


def read_pdf_file_return_pages_list(file_path):
    reader=PyPDF2.PdfReader(file_path)
    pages=[]
    for page in reader.pages:
        pages.append(page.extract_text())
    return pages

def claude_translate_from_hebrew(text):
    res=_claude.messages.create(
        model="claude-3-opus-20240229",
        max_tokens=1000,
        temperature=0.0,
        system="translate the input into English and return the translation",
        
        messages=[{"role": "user", "content": text}]
    )
    return res
def groq_translate_from_hebrew(text):
    chat_completion = _groq.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": "translate the following text into English :"+text
            }
        ],
        model="mixtral-8x7b-32768"
        # model="llama3-70b-8192",
    )
    return chat_completion.choices[0].message.content
```

```python
pages=read_pdf_file_return_pages_list('Cleaning_SOP.pdf')
print(pages[0])
res=claude_translate_from_hebrew(pages[0])
print(res.content[0].text)
```
