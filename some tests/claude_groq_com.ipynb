```python
import anthropic
import os
import PyPDF2

def read_pdf_file_return_pages_list(file_path):
    reader=PyPDF2.PdfReader(file_path)
    pages=[]
    for page in reader.pages:
        pages.append(page.extract_text())
    return pages
```
