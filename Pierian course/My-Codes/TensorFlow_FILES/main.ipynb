```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('../TensorFlow_FILES/DATA/fake_reg.csv')

```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['feature1','feature2']].values
y=df['price'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
#plot the x_train
plt.scatter(X_train[:,0],y_train,color='red',label='X_train')
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

scaler.fit(X_train)

X_train=scaler.transform(X_train)

X_test=scaler.transform(X_test)
```

```python
#plot the x_train
plt.scatter(X_train[:,0],y_train,color='red',label='X_train')
```

```python
#plot the x_train before and after scaling in the same plot
plt.scatter(X_train[:,0],y_train,color='red',label='X_train')
plt.scatter(X_test[:,0],y_test,color='blue',label='X_test')
```
